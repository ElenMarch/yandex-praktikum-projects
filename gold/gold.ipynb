### Проверка расчетов эффективности обогащения

* C — доля золота в концентрате после флотации/очистки;
* F — доля золота в сырье/концентрате до флотации/очистки;
* T — доля золота в отвальных хвостах после флотации/очистки.

Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах.

```python
def recovery(row):
    f = row['rougher.input.feed_au']
    c = row['rougher.output.concentrate_au']
    t = row['rougher.output.tail_au'] 

    return (c*(f-t))*100/(f*(c-t))
```

```python
gold_train['recovery'] = gold_train.apply(recovery, axis=1)
```

```python
gold_train.loc[:, ['recovery', 'rougher.output.recovery']].head()
```

```python
mae = mean_absolute_error(gold_train['rougher.output.recovery'], gold_train['recovery'])
mae
```

```python
gold_train = gold_train.drop(['recovery'], axis=1)
```

## Модель

```python
def sMAPE(target, predict):
    smape = (200*abs(target-predict) / (abs(target) + abs(predict))).mean()
    return 0.25 *smape[0] + 0.75*smape[1]
```

```python
sm_score = make_scorer(sMAPE, greater_is_better = False)
```

Добавим в тестовую выборку целевые признаки `rougher.output.recovery`, `final.output.recovery`.

```python
features_test = gold_test.reset_index(drop=True)
target_test = gold_full.loc[gold_test.index, ['rougher.output.recovery', 'final.output.recovery']].reset_index(drop=True)
```

```python
features_test.shape, target_test.shape
```

```python
features = gold_train_clean[gold_test.columns].reset_index(drop=True)
target = gold_train_clean[['rougher.output.recovery','final.output.recovery']].reset_index(drop=True)
```

```python
features.shape, target.shape
```

```python
scaler = StandardScaler()
scaler.fit(features)
features = scaler.transform(features)
features_test= scaler.transform(features_test)
```

Для решения задачи регрессии подберем подходящую модель, сравним ее с константной:

```python
cv_models = pd.DataFrame()
```

```python
logic_model = LinearRegression()
score_lr = cross_val_score(logic_model, features, target, scoring= sm_score, cv=5).mean()
print("Наилучшая модель Линейной регрессии имеет оценку:", abs(score_lr))
```

```python
cv_models = pd.DataFrame({'sMape': abs(score_lr), 'params': [None], 'model_name': "Линейная регрессия"}, index=[0])
```

```python
tree_model = DecisionTreeRegressor(random_state=147)
params= {'min_samples_split': range(2,6),
         'max_depth': range(1,10),
         'min_samples_leaf': range(1,4)}

tree_grid = RandomizedSearchCV(tree_model, params, scoring= sm_score, cv=5)
tree_grid.fit(features, target)
print("Наилучшая модель для решающего дерева: %f с параметрами %s" %(abs(tree_grid.best_score_), tree_grid.best_params_))
```

```python
cv_models = pd.concat([cv_models,pd.DataFrame({'sMape': abs(tree_grid.best_score_), 'params': [tree_grid.best_params_],
                                               'model_name': "DecisionTree"}, index=[1])])
```

```python
forest_model = RandomForestRegressor(random_state=147)
param = {'n_estimators': range(60, 101, 20),
         'max_depth': range(1, 7)}

forest = RandomizedSearchCV(forest_model, param, scoring=sm_score, cv=5)
forest.fit(features, target)
print("Наилучшая модель для случайного леса: %f с параметрами %s" % (abs(forest.best_score_), forest.best_params_))
```
